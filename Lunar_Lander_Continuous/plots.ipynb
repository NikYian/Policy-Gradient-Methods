```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 300

RF_scores = np.load('REINFORCE/REINFORCE_scores.npy')
RF_ra = np.convolve(RF_scores, np.ones(N)/N, mode='valid') 
RF_bs_scores = np.load('REINFORCE_Baseline/RF_bsl_scores.npy')
RF_bs_ra = np.convolve(RF_bs_scores, np.ones(N)/N, mode='valid') 
AC_scores = np.load('Actor_Critic/ACTOR_CRITIC_scores.npy')
AC_ra = np.convolve(AC_scores, np.ones(N)/N, mode='valid') 
```

```python
fig, ax = plt.subplots(figsize = (10,4))
# plt.plot(RF_scores, alpha=0.2, color = 'seagreen')
# plt.plot(RF_bs_scores, alpha=0.2, color= 'salmon')
# plt.plot(AC_scores, alpha=0.2, color= 'blue')
ax.plot(AC_ra, label = 'Actor Critic', color= 'blue')
ax.plot(RF_bs_ra, label = 'REINFORCE with Baseline', color= 'salmon')
ax.plot(RF_ra, label = 'REINFORCE', color = 'seagreen')


plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# plt.title('Policy Collapse',fontsize=14)
ax.legend(fontsize=14)
ax.set_xlabel('Episode', fontsize=14)
ax.set_ylabel('Score', fontsize=14)
plt.show()
```
