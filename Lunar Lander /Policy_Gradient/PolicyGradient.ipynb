```python
def load_checkpoint(agent, checkpoint_path):
    if os.path.exists(checkpoint_path):
        print(f"Loading checkpoint from {checkpoint_path}")
        agent.policy.load_state_dict(torch.load(checkpoint_path))
    else:
        print(f"No checkpoint found at {checkpoint_path}")

%load_ext autoreload
%autoreload 2
```

# Agent training

```python
import gym
import imageio
from Agent import PolicyGradientAgent
import torch
import os
import matplotlib.pyplot as plt
import numpy as np



if __name__ == '__main__':
    env = gym.make('LunarLander-v2')

    n_games = 5000
    obs = env.reset()

    agent = PolicyGradientAgent(lr=5e-4, input_dims=obs.shape)

    # Load a checkpoint if available
    checkpoint_path = "checkpoint.pth"  # Update with your desired checkpoint
    load_checkpoint(agent, checkpoint_path)

    score_ls = []

    checkpoint_interval = 100

    for i in range(2900,n_games):
        obs = env.reset()

        score = 0
        done = False
        while not done:
            action = agent.choose_action(obs)
            obs, reward, done,  info = env.step(action)
            score += reward
            agent.store_reward(reward)

        agent.learn()
        score_ls.append(score)
        avg_score = np.mean(score_ls[-100:])
        if (i%100 == 0):
            print('episode', i, 'avg_score %.1f' %avg_score)

        if i % checkpoint_interval == 0 and i > 0:
            checkpoint_path = f"checkpoint.pth"
            torch.save(agent.policy.state_dict(), checkpoint_path)
            print(f"Checkpoint saved at episode {i}")

        if (i%100 == 0):
            print('episode', i, 'avg_score %.1f' %avg_score)

    # Save the final model after training
    torch.save(agent.policy.state_dict(), checkpoint_path)
    env.close()
```
